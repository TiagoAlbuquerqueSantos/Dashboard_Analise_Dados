```python
colunas_traduzidas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df.rename(columns=colunas_traduzidas, inplace=True)

print(df.columns)
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

df['contrato'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
})

df['tamanho_empresa'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

df['remoto'].value_counts()
```

```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['João', 'Maria', 'Pedro', 'Ana', 'José'],
    'salario': [4000, np.nan, 5000, np.nan, 100000],
})

# Calcula a média e faz o aredondamento
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Calcula a mediana e faz a substituição
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temps = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27],
})

df_temps['Preenchido_ffill'] = df_temps['temperatura'].ffill()

df_temps
```

```python
df_temps = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27],
})

df_temps['Preenchido_bfill'] = df_temps['temperatura'].bfill()

df_temps
```

```python
df_cidades = pd.DataFrame({
    'nome': ['João', 'Maria', 'Pedro', 'Ana', 'José'],
    'cidade': ['São Paulo', np.nan, 'Belo Horizonte', np.nan, 'Belém'],
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')

df_cidades
```

```python
import plotly.express as px
```

```python
df_grouped = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()
fig = px.bar(
    df_grouped,
    x='senioridade',
    y='usd',
    title='Média Salarial por Senioridade',
    labels={'senioridade': 'Senioridade', 'usd': 'Média Salarial em USD'})
fig.show()
```

```python
remoto_cont = df_limpo['remoto'].value_counts().reset_index()
remoto_cont.columns = ['Tipo_trabalho', 'quantidade']
fig = px.pie(
    remoto_cont,
    names='Tipo_trabalho',
    values='quantidade',
    title='Proporção dos tipos de trabalho',
    hole=0.5
    )
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
import pycountry

def converter_sigla_pais_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(converter_sigla_pais_iso3)

df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

fig = px.choropleth(
    media_ds_pais,
    locations='residencia_iso3',
    color='usd',
    color_continuous_scale='Viridis',
    title='Salário médio de Cientista de Dados por País',
    labels={'usd': 'Salário Médio em USD', 'residencia_iso3': 'Pais'}
)

fig.show()
```
